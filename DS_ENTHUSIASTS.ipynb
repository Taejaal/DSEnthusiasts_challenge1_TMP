ii) Categorical analysis

```python
ignore_cols = ['Customer_ID', 'Transaction_ID', 'Phone', 'Email', 'Name', 'Address', 'Zipcode', 'Date', 'Time','Month']

categorical_cols = [col for col in df.select_dtypes(include=['object']).columns if col not in ignore_cols]
print("Categorical Columns:")
for i, col in enumerate(categorical_cols, 1):
    print(f"{i}. {col}")
```

```python
# Load dataset (adjust path as needed)
df = pd.read_csv("retail_data.csv")

sns.set(style="whitegrid", palette="flare")

# --- Safe bar plot function ---
def safe_bar(ax, data, title, colors):
    ax.bar(range(len(data)), data.values, color=colors)
    ax.set_xticks(range(len(data)))
    ax.set_xticklabels(data.index, rotation=45, ha='right')
    ax.set_title(title)
    ax.set_ylabel("Count")

# --- Pie chart function ---
def safe_pie(ax, data, title):
    ax.pie(data.values, labels=data.index, autopct='%1.1f%%', startangle=90)
    ax.axis('equal')
    ax.set_title(title)

# --- Color Palettes ---
colors_oranges = sns.color_palette("Oranges", 10)
colors_reds = sns.color_palette("Reds", 10)

# --- Plots to Generate ---
plots = [
    ('City', "Top 10 Cities", 'bar', colors_oranges),
    ('Gender', "Gender Distribution", 'bar', colors_reds),
    ('Income', "Income Distribution", 'pie', None),
    ('Customer_Segment', "Customer Segment", 'bar', colors_oranges),
    ('Month', "Month-wise Distribution", 'bar', colors_reds),
    ('Product_Category', "Product Category", 'bar', colors_oranges),
    ('Product_Brand', "Top 10 Product Brands", 'bar', colors_reds),
    ('Product_Type', "Top 10 Product Types", 'bar', colors_oranges),
    ('Feedback', "Feedback", 'pie', None),
    ('Shipping_Method', "Shipping Method", 'bar', colors_oranges),
    ('Payment_Method', "Payment Method", 'pie', None),
    ('Order_Status', "Order Status", 'bar', colors_reds),
]

# Prepare dynamic grid
num_plots = len(plots)
cols = 3
rows = (num_plots + cols - 1) // cols

fig, axes = plt.subplots(rows, cols, figsize=(20, 5 * rows))
fig.suptitle("Univariate Analysis of Categorical Columns", fontsize=18)
axes = axes.flatten()

# Loop through plots
for i, (col, title, plot_type, palette) in enumerate(plots):
    ax = axes[i]

    # Special case for months (to preserve order)
    if col == 'Month':
        month_order = ['January','February','March','April','May','June',
                       'July','August','September','October','November','December']
        data = df[col].value_counts().reindex(month_order).fillna(0)
    elif col in ['Product_Brand', 'Product_Type', 'City']:
        data = df[col].value_counts().nlargest(10)
    else:
        data = df[col].value_counts()

    # Plot
    if plot_type == 'bar':
        safe_bar(ax, data, title, palette)
    elif plot_type == 'pie':
        safe_pie(ax, data, title)

# Remove empty plots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

a) Scatter plot

```python
%matplotlib inline

df.columns = df.columns.str.strip().str.lower()

kpi_pairs = [
    ('age', 'amount'),
    ('age', 'total_purchases'),
    ('age', 'total_amount'),
    ('total_purchases', 'total_amount'),
    ('amount', 'total_amount'),
    ('total_purchases', 'ratings')
]

for col in set([c for pair in kpi_pairs for c in pair]):
    df[col] = pd.to_numeric(df[col], errors='coerce')

cols = 3
rows = 2
plt.figure(figsize=(cols*5, rows*4))

for i, (x_col, y_col) in enumerate(kpi_pairs, 1):
    plot_df = df[[x_col, y_col]].dropna()

    plt.subplot(rows, cols, i)
    sns.scatterplot(data=plot_df, x=x_col, y=y_col, color='orange')

    if not plot_df.empty:
        corr = plot_df.corr().iloc[0,1]
        plt.title(f'{x_col} vs {y_col}\nCorr: {corr:.2f}')
    else:
        plt.title(f'{x_col} vs {y_col}\nNo data')

    plt.xlabel(x_col)
    plt.ylabel(y_col)

plt.tight_layout()
plt.show()
```

Observations:

Age vs Amount / Total_Purchases / Total_Amount – Middle-age groups often cluster at moderate–high spending and purchase levels.

Total_Purchases vs Total_Amount – Strong positive correlation; total spend rises mainly with purchase frequency.

Amount vs Total_Amount – Strong correlation; larger single purchases contribute to higher overall spend.

Total_Purchases vs Ratings – Frequent buyers tend to give slightly higher ratings.

Outliers – A few customers with very high amounts or purchases stand apart; could indicate VIPs or data anomalies.

b) Regression line

```python
%matplotlib inline

df.columns = df.columns.str.strip().str.lower()

kpi_pairs = [
    ('age', 'amount'),
    ('age', 'total_purchases'),
    ('age', 'total_amount'),
    ('total_purchases', 'total_amount'),
    ('amount', 'total_amount'),
    ('total_purchases', 'ratings')
]

for col in set([c for pair in kpi_pairs for c in pair]):
    df[col] = pd.to_numeric(df[col], errors='coerce')

cols = 3
rows = 2
plt.figure(figsize=(cols*5, rows*4))

for i, (x_col, y_col) in enumerate(kpi_pairs, 1):
    plot_df = df[[x_col, y_col]].dropna()

    plt.subplot(rows, cols, i)
    sns.regplot(data=plot_df, x=x_col, y=y_col, scatter=False, color='orange', line_kws={'linewidth':2})

    if not plot_df.empty:
        corr = plot_df.corr().iloc[0,1]
        plt.title(f'{x_col} vs {y_col}\nCorr: {corr:.2f}')
    else:
        plt.title(f'{x_col} vs {y_col}\nNo data')

    plt.xlabel(x_col)
    plt.ylabel(y_col)

plt.tight_layout()
plt.show()
```

Observations:

1. **Age vs Amount** – Middle-aged customers generally spend more per purchase; slight upward trend.
2. **Age vs Total\_Purchases** – Purchase frequency increases moderately with age, peaking at middle age.
3. **Age vs Total\_Amount** – Total spend rises with age initially, then flattens for older groups.
4. **Total\_Purchases vs Total\_Amount** – Strong positive trend; more purchases → higher total spend.
5. **Amount vs Total\_Amount** – Larger single purchases contribute to higher overall revenue.
6. **Total\_Purchases vs Ratings** – Frequent buyers show slightly higher ratings; trend is mild.

# 8. OUTLIERS ANALYSIS

Identify outliers using the Interquartile Range (IQR) method


```python
numerical_cols = ['Transaction_ID', 'Customer_ID', 'Phone', 'Zipcode', 'Age', 'Year', 'Total_Purchases', 'Amount', 'Total_Amount', 'Ratings']

outliers_iqr = {}
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    outliers_iqr[col] = outliers

print("Outliers identified using IQR method:")
for col, outliers in outliers_iqr.items():
    print(f"\nOutliers in '{col}':")
    if not outliers.empty:
        display(outliers)
    else:
        print("No outliers found.")
```

```python
print("Descriptive statistics for outliers:")

for col, outliers in outliers_iqr.items():
    if not outliers.empty:
        print(f"\nDescriptive statistics for outliers in '{col}':")
        display(outliers.describe())
```
